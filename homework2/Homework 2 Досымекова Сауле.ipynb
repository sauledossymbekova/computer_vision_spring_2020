## Problem 2

a) Blur your image with kernel size 5x5, 10x10 and 2x5. Show the result.

b) Blur half part of your image (kernal size 15x15). Show the result.

c) Take any piece of the image.Blur this piece and select this piece in the image in the green frame. Show the result.

```python
image=cv2.imread('leaf.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
K_1 = np.ones((5,5),np.float32)/(5*5)
K_2 = np.ones((10,10),np.float32)/(10*10)
K_3 = np.ones((2,5),np.float32)/(2*5)

image_bl1 = cv2.filter2D(image,-1,K_1)
image_bl2 = cv2.filter2D(image,-1,K_2)
image_bl3 = cv2.filter2D(image,-1,K_3)

plt.imshow(image),plt.title('Root')
```

```python
plt.imshow(image_bl1),plt.title('5*5')
```

```python
plt.imshow(image_bl2),plt.title('10*10')
```

```python
plt.imshow(image_bl3),plt.title('2*5')
```

```python
image=cv2.imread('leaf.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
K = np.ones((15,15),np.float32)/(15*15)
image_b = cv2.filter2D(image,-1,K)

plt.subplot(121),plt.imshow(image),plt.title('Root')
plt.subplot(122),plt.imshow(image_b),plt.title('15*15')
plt.xticks([0,128]), plt.yticks([128,256])
plt.show()
```

```python
image=cv2.imread('leaf.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

K = np.ones((15,15),np.float32)/(15*15)
image_c = cv2.filter2D(image,-1,K)

image_ = cv2.rectangle(image, (1,128),(255,255),(0,255,0),4)
plt.subplot(121),plt.imshow(image_),plt.title('Root')
plt.subplot(122),plt.imshow(image_c),plt.title('15*15')
plt.xticks([0,128]), plt.yticks([128,256])
plt.show()
```

```python
image=cv2.imread('leaf.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

K = np.ones((15,15),np.float32)/(15*15)

image[128:255,1:255]=cv2.filter2D(image[128:255,1:255],-1,K)

cv2.rectangle(image,(1,128),(255,255),(0,255,0),4)
plt.imshow(image)
```

Create a randomly generated slice inside your picture which will be blurred and grayscale

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import cv2
import random
image=cv2.imread('leaf.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python

n1=random.randint(0,255)
n2=random.randint(n1,255)
n3=random.randint(0,255)
n4=random.randint(n3,255)

gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

K=np.ones((5,5),np.float32)/(5*5)

image[n1:n2,n3:n4,0]=cv2.filter2D(gray[n1-1:n2-1,n3-1:n4-1],-1,K)
image[n1:n2,n3:n4,1]=cv2.filter2D(gray[n1-1:n2-1,n3-1:n4-1],-1,K)
image[n1:n2,n3:n4,2]=cv2.filter2D(gray[n1-1:n2-1,n3-1:n4-1],-1,K)

plt.imshow(image)
```
