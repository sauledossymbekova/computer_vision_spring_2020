## Image Fourier Transform
___

[Canny edge detector](https://en.wikipedia.org/wiki/Canny_edge_detector)

[OpenCV](https://docs.opencv.org/trunk/da/d22/tutorial_py_canny.html)

[Canny](https://docs.opencv.org/trunk/dd/d1a/group__imgproc__feature.html#ga04723e007ed888ddf11d9ba04e2232de)

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('org.png',0)
edges = cv2.Canny(img,10,10)

plt.figure()

plt.subplot(1,2,1)
plt.imshow(img)

plt.subplot(1,2,2)
plt.imshow(edges)

plt.show()
```

## Hough Tranform
___
[OpenCV](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_houghlines/py_houghlines.html)

```python
img = cv2.imread('org.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,10,10)

lines = cv2.HoughLinesP(edges,1,np.pi/180,100)
for i in range(len(lines)):
    for x1,y1,x2,y2 in lines[i]:
        print(x1, y1, x2, y2)
        cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)

plt.imshow(img)
```
