## Image Fourier Transform
___

[Fourier Transform](https://en.wikipedia.org/wiki/Fourier_transform)

[OpenCV](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_transforms/py_fourier_transform/py_fourier_transform.html)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('org.png',0)
f = np.fft.fft2(img)
# fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(f))

plt.figure()

plt.subplot(1,2,1)
plt.imshow(img)

plt.subplot(1,2,2)
plt.imshow(magnitude_spectrum)

plt.show()
```

```python
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

plt.figure()

plt.subplot(1,2,1)
plt.imshow(img)

plt.subplot(1,2,2)
plt.imshow(magnitude_spectrum)

plt.show()
```

```python
f = np.fft.fft2(np.rot90(img))
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))

plt.figure()

plt.subplot(1,2,1)
plt.imshow(np.rot90(img))

plt.subplot(1,2,2)
plt.imshow(magnitude_spectrum)

plt.show()
```
