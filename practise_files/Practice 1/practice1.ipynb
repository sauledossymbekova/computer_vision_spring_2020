# Image procesing
___

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
```

## Average
\begin{equation}
I = \frac{(R + G + B)}{3}
\end{equation}

```python
I = (R + G + B)/3
plt.imshow(I)
```

```python
plt.figure(figsize=(15, 15))

# Original image
plt.subplot(1, 2, 1)
plt.title('Original image')
plt.imshow(img)

# Grayscale image
plt.subplot(1, 2, 2)
plt.title('Grayscale image')
plt.imshow(I)

plt.show()
```

## Luminosity
\begin{equation}
I = 0.21*R + 0.72*G + 0.07*B
\end{equation}

```python
I = 0.21 * R + 0.72 * G + 0.07 * B
plt.imshow(I)
```

```python
plt.figure(figsize=(15, 15))

# Original image
plt.subplot(1, 2, 1)
plt.title('Original image')
plt.imshow(img)

# Grayscale image
plt.subplot(1, 2, 2)
plt.title('Grayscale image')
plt.imshow(I)

plt.show()
```

```python
cv2.imwrite('test.png',I)
```
